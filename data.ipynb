```python
#IQR function
def outliers(df,colomne):
   Q1=df[colomne].quantile(0.25)
   Q3=df[colomne].quantile(0.75)
   IQR=Q3-Q1
   outlier_indices=df[(df[colomne]<(Q1-1.5*IQR))| (df[colomne]>(Q3+1.5*IQR))].index
   return outlier_indices
outlier_indices=outliers(df,"person_income")
#convert outlier_indices
outlier_indices=pd.DataFrame(outlier_indices)
outlier_indices
```
